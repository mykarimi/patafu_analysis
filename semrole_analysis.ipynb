```python
sns.set(color_codes=True)
sns.set_context
np.random.seed(sum(map(ord, "categorical")))
a_semrole = sns.load_dataset("a_semrole")
sns.swarmplot(x="Semantic Role", y="F-measure", hue="Evaluation", data=a_semrole, size=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
print('INTRA-PERIODIC EVENT COREFERENCE EVALUATION OF ONE EVENTINSTANCE')
sns.plt.show()
```
